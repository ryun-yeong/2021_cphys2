```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
train = pd.read_csv('/Users/gimlyun-yeong/Desktop/titanic/train.csv')
test = pd.read_csv('/Users/gimlyun-yeong/Desktop/titanic/test.csv')
```

```python
train_test_data = [train, test]

for dataset in train_test_data:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
#dataset의 Name에서 성별, 직업 등과 관련된 Mr, Miss, Mrs 등으로 Title이라는 새로운 열을 만든다.
```

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].map(title_mapping)
#Mr = 0, Miss = 1, Mrs = 2 그 외에 나머지 Title값은 3을로 바꾼다.
```

```python
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)
#Title값에 넣은 정보를 빼면 의미가 없을 듯 하여 삭제한다.
```

```python
sex_mapping = {"male": 0, "female": 1}
for dataset in train_test_data:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
#남성이며 0, 여성이면 1로 즉, 숫자로 바꾼다.
```

```python
train["Age"].fillna(train.groupby("Title")["Age"].transform("median"), inplace=True)
test["Age"].fillna(test.groupby("Title")["Age"].transform("median"), inplace=True)
#Age에서 비어있는 값의 age를 Title이 같은 data들의 중간값으로 사용한다.
```

```python
for dataset in train_test_data:
    dataset['Age'] = dataset['Age']/9.5
#나이가 가장 많은 사람이 76살인데 나이는 생존에 많은 영향을 주었을 것으로 예상하여 38로 9.5로 나누어 0~8까지의 데이터로 만든다.
```

```python
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
#Embarked data에서 S가 매우 많아 비어있는 곳을 S로 채운다.
```

```python
embarked_mapping = {"S": 0, "C": 1, "Q": 2}
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].map(embarked_mapping)
# embarked data도 숫자로 바꾸어 준비한다.
```

```python
train["Fare"].fillna(train.groupby("Pclass")["Fare"].transform("median"), inplace=True)
test["Fare"].fillna(test.groupby("Pclass")["Fare"].transform("median"), inplace=True)
train.head()
#Fare의 값이 비어있는 경우 같은 Pclass의 Fare 값 중 중간값을 사용한다.
```

```python
np.max(dataset['Fare'])
```

```python
for dataset in train_test_data:
    dataset['Fare'] = dataset['Fare']/(64)
#타이타닉의 1등급호실이 제일 늦게 가라앉았으니 영향이 클 것이라 생각하고 0~8까지의 data로 바꾸었다.
```

```python
for dataset in train_test_data:
    dataset['Cabin'] = dataset['Cabin'].str[:1]
#Cabin에서 알파벳 뒤의 숫자를 제거했다.
```

```python
cabin_mapping = {"A": 0, "B": 0.4, "C": 0.8, "D": 1.2, "E": 1.6, "F": 2, "G": 2.4, "T": 2.8}
for dataset in train_test_data:
    dataset['Cabin'] = dataset['Cabin'].map(cabin_mapping)
#Cabin에서 알파벳을 숫자로 바꾸어 준비했다.
```

```python
train["Cabin"].fillna(train.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
test["Cabin"].fillna(test.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
#숫자로 바꾼 Cabin값들에서 빠진 값들은 같은 Pclass의 Cabin값의 중간 값을 사용했다.
```

```python
train["Size"] = train["SibSp"] + train["Parch"] + 1
test["Size"] = test["SibSp"] + test["Parch"] + 1
#SibSp와 Parch는 둘 다 동승자의 숫자를 뜻하니 Size로 한 번에 합쳐서 사용했다.
```

```python
Size_mapping = {1: 0, 2: 0.5, 3: 1.0, 4: 1.5, 5: 2.0, 6: 2.5, 7: 3.0, 8: 3.5, 9: 4.0, 10: 4.5, 11: 5.0}
for dataset in train_test_data:
    dataset['Size'] = dataset['Size'].map(Size_mapping)
#동승자의 숫자를 mapping으로 0~5까지의 숫자로 바꾸었다.(동승자의 숫자도 꽤 영향을 끼쳤을 것으로 예상)
```

```python
for dataset in train_test_data:
    dataset['Sex'] = dataset['Sex']*2
for dataset in train_test_data:
    dataset['Title'] = dataset['Title']*4
#성별의 영향이 꽤 있을 것 같아서 0~2까지의 data로 만들고 성별이 또 포함되긴 한 Title을 0~12까지로 바꾸었다.
```

```python
features_drop = ['Ticket', 'SibSp', 'Parch']
train = train.drop(features_drop, axis=1)
test = test.drop(features_drop, axis=1)
train = train.drop(['PassengerId'], axis=1)
#Fare가 있으니 Ticket을 없애고 Size로 묶인 SibSp와 Parch도 없앤다. 그리고 필요없는 PassengerId를 없앴다.
```

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']

train_data.shape, target.shape
#Survived를 뺀 데이터를 train_data로 Survived만 남은 data를 target으로 설정했다.
```

```python
train_data.head(10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import Perceptron

import numpy as np
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
#data를 10개로 나누어서 validation을 하겠다.
```

```python
clf = KNeighborsClassifier(n_neighbors = 14) #KNN의 neighbor를 14개를 사용한다.
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
#10개로 나누어서 validation한 정확도를 각각 출력한다.
```

```python
round(np.mean(score)*100, 2)
#crossvalidation결과를 합쳐서 평균을 낸다.
```

```python
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
#KNN과 동일한 방식으로 DecisionTreeClassifier를 사용함
```

```python
# decision tree Score
round(np.mean(score)*100, 2)
```

```python
clf = Perceptron()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
#위와 동일한 방법으로 Perceptron사용
```

```python
round(np.mean(score)*100,2)
```
